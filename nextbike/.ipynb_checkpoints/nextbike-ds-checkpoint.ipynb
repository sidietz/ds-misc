```python
import pandas as pd
```

```python
nb = pd.read_csv("nb.csv")
```

```python
count = pd.read_csv("count.csv")
```

```python
count = count[(count != 0).all(1)]
```

```python
nxb = pd.merge(nb, count, how="inner")
```

```python
p2 = pd.DataFrame([["Potsdam 2", 175710, 932, 486, 44, "Präsi"], ["Potsdam 1", 175710, 932, 386, 44, "Präsi"]], columns=list(nxb.columns))
```

```python
nxb = nxb.append(p2)
```

```python
def to_factor(x):
    #f = round(x / 500.0) * 500.0
    f = (round(x, -3)/1000)
    return 1/f
```

```python
nxb['factor'] = nxb.apply(lambda x: to_factor(x.density), axis=1)
```

```python
nxb['h_per_bike'] = nxb.apply(lambda x: x.population / x.bikes, axis=1)
```

```python
nxb['h_per_bike&density'] = nxb.apply(lambda x: (x.factor * x.h_per_bike), axis=1)
```

```python
nextbike = nxb[['city', 'population', 'density', 'bikes', 'stations', 'h_per_bike', 'h_per_bike&density']].sort_values("h_per_bike&density").reset_index(drop=True)
```
