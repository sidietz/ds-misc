```python
def get_tree(city):
    url = str("http://de.wikipedia.org/wiki/" + str(city))
    return get_city(url)

def get_number(alist, name):
    #idx = alist.index(name)
    indices = [i for i, s in enumerate(alist) if name in s]
    try:
        idx = indices[0]
    except IndexError:
        return 0
    found = False
    while not found:
        idx+=1
        try:
            found, res = match_inhabs(alist[idx].strip(" "))
        except IndexError:
            return 0
    return res


def match_inhabs(s):
    match = re.findall(r"\d", s)
    try:
        result = int("".join(match))
        return True, result
    except:
        return False, 0

    
def get_table_list(tree, rx="//tr//text()"):
    aa = tree.xpath(rx)
    return aa


def aq_pop(city):
    tree = get_tree(city)
    alist = get_table_list(tree)
    return get_number(alist, "Einwohner")


def aquire_data(city, name):
    tree = get_tree(city)
    alist = get_table_list(tree)
    num = get_number(alist, name)
    return num
```

```python
def get_cities(txt):
    match = re.findall(r"[A-Z]\w+", txt)
    result = list(set(match))
    result.sort()
    return result
```

```python
def get_city(url):
    r = rq.get(url, headers=HEADER)
    text = r.text
    # print(text)
    #url_string = bytes(r.text, 'iso-8859-1').decode('utf-8')
    tree = html.document_fromstring(r.text)
    return tree
```

```python
nb["population"] = nb.apply(lambda x: aquire_data(x.city, "Einwohner"), axis=1)
```

```python
nb["density"] = nb.apply(lambda x: aquire_data(x.city, "Bevölkerungsdichte"), axis=1)
```
